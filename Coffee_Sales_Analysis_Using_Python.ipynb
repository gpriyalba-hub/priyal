○ At what time of the day are people most likely to buy coffee?

```python
# --- Count purchases by hour ---
hourly_sales = df['hour'].value_counts().sort_index()

# --- Display hourly purchase data ---
print("Number of Coffee Purchases by Hour of Day:")
print(hourly_sales)

# --- Plot hourly trends ---
plt.figure(figsize=(10,5))
plt.plot(hourly_sales.index, hourly_sales.values, marker='o', linestyle='-', linewidth=2, color='teal')
plt.title("Coffee Purchases by Time of Day", fontsize=16, fontweight='bold')
plt.xlabel("Hour of the Day (24-hour format)", fontsize=12)
plt.ylabel("Number of Purchases", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# --- Highlight peak hour ---
peak_hour = hourly_sales.idxmax()
peak_sales = hourly_sales.max()
plt.annotate(f"Peak: {peak_hour}:00 ({peak_sales} sales)",
             xy=(peak_hour, peak_sales),
             xytext=(peak_hour+1, peak_sales+30),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=10, color='red')

plt.show()
```

# Insights — Coffee Purchases by Time of Day:

- The analysis shows that coffee sales peak between 10 AM and 12 PM, with 10:00 AM being the busiest hour of the day.

- This pattern indicates that most customers buy coffee during mid-morning hours, likely as a boost before or during work.

- Sales begin increasing after 7 AM, remain high until around 2 PM, and then gradually decline toward evening.

- Very few purchases occur after 8 PM, suggesting that coffee is mainly a morning-to-afternoon beverage for most customers.

- This trend helps businesses plan better — for example:

   -Schedule more staff or faster preparation during peak hours (10–12 AM).
   
  -Offer discounts or promotions in the quieter evening hours to attract more customers.

# 2. Sales Analysis:

○ Which coffee earns the most revenue?

```python
# Clean column names (remove spaces if any)
df.columns = df.columns.str.strip()


#  Calculate revenue per coffee

revenue_per_coffee = (
    df.groupby('coffee_name')['money']
    .sum()
    .reset_index()
    .sort_values(by='money', ascending=False)
)

# Display top 5 revenue-generating coffees
print("Top 5 Coffees by Revenue:")
print(revenue_per_coffee.head())

#  Plot: Coffee Revenue Comparison

plt.figure(figsize=(10,6))
plt.bar(revenue_per_coffee['coffee_name'], revenue_per_coffee['money'],
        color='peru', edgecolor='black')
plt.title('Coffee Revenue Comparison', fontsize=16, fontweight='bold')
plt.xlabel('Coffee Type', fontsize=12)
plt.ylabel('Total Revenue (₹)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Highlight top coffee
top_coffee = revenue_per_coffee.iloc[0]
plt.annotate(f"Top: {top_coffee['coffee_name']} ({top_coffee['money']:.2f} ₹)",
             xy=(0, top_coffee['money']),
             xytext=(1, top_coffee['money'] + 1000),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=10, color='darkred')

plt.show()
```

# Insights — Coffee Revenue Analysis:

- Latte generates the highest total revenue at ₹28,658.30, making it the top.

- It’s followed by Americano with Milk, Cappuccino, Americano, and Hot Chocolate.

- The chart shows a clear dominance of Latte and milk-based coffees, suggesting customer preference for richer, creamier drinks.

- Focusing promotions or premium variants around these top 3 could maximize sales revenue.

○ What is the average ticket size for different coffees?

```python
# Clean column names
df.columns = df.columns.str.strip()

# Compute average ticket size per coffee
avg_ticket_size = (
    df.groupby('coffee_name')['money']
    .mean()
    .reset_index()
    .sort_values(by='money', ascending=False)
)

# Plot
plt.figure(figsize=(10,6))
bars = plt.bar(avg_ticket_size['coffee_name'], avg_ticket_size['money'],
               color='sandybrown', edgecolor='black')

plt.title('Average Ticket Size per Coffee Type', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Coffee Type', fontsize=12)
plt.ylabel('Average Ticket Size (₹)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout(rect=[0, 0, 1, 0.98])  # make space for title

# Identify the top coffee
top_index = avg_ticket_size['money'].idxmax()
top_value = avg_ticket_size.loc[top_index, 'money']
top_name = avg_ticket_size.loc[top_index, 'coffee_name']

# Add a clear annotation (inside the plot area)
plt.annotate(f"Highest Avg: {top_name}\n₹{top_value:.2f}",
             xy=(list(avg_ticket_size.index).index(top_index), top_value),
             xytext=(list(avg_ticket_size.index).index(top_index), top_value + 2),
             ha='center',
             fontsize=10,
             color='darkred',
             arrowprops=dict(facecolor='black', arrowstyle='->', lw=1))

plt.show()
```

# Insights — Average Ticket Size per Coffee:

- The average ticket size (average amount spent per order) is highest for Hot Chocolate (₹36.07), followed closely by Cappuccino, Cocoa, and Latte.

- These drinks tend to be premium or milk-heavy beverages, explaining their higher price points.

- Double Espresso with Milk also appears among the top, indicating customers are willing to pay more for richer, milk-based or chocolate-flavored drinks.

- This analysis helps in pricing strategy — maintaining margins on high-value coffees and offering combos or loyalty offers around these popular items.

○ How do daily sales vary over time?

```python
# Clean and prepare data
df.columns = df.columns.str.strip()
df['date'] = pd.to_datetime(df['date'])
df['money'] = pd.to_numeric(df['money'], errors='coerce')


#  Calculate Daily Total Sales

daily_sales = df.groupby('date')['money'].sum().reset_index()
daily_sales = daily_sales.sort_values(by='date')

# Add a 7-day moving average for trend smoothing
daily_sales['7_day_avg'] = daily_sales['money'].rolling(window=7).mean()

# Display sample data
print(" Daily Sales Summary:")
print(daily_sales.head(10))


#  Plot: Daily Sales Trend Over Time

plt.figure(figsize=(12,6))

# Line for daily sales
plt.plot(daily_sales['date'], daily_sales['money'],
         marker='o', linestyle='-', color='teal', label='Daily Sales')

# Line for 7-day moving average
plt.plot(daily_sales['date'], daily_sales['7_day_avg'],
         color='orange', linewidth=3, linestyle='--', label='7-Day Moving Avg')

# Chart styling
plt.title("Daily Coffee Sales Trend Over Time", fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Total Sales (₹)", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout(rect=[0, 0, 1, 0.98])

# Display the plot
plt.show()
```

# Insights — Daily Sales Trend Analysis:

- The daily sales fluctuate noticeably over the period, showing peaks and dips across different dates.

- This variation could be due to weekend effects, special offers, or changing customer traffic.

- For example, March 1st shows relatively higher total sales, suggesting it could be a busy or promotional day.

- Overall, the trend line helps identify consistent sales days vs. low-performing days, useful for scheduling staff or planning discounts.

○ Is there a noticeable difference in customer behavior between February and March?

```python
df.columns = df.columns.str.strip()
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month_name()
df['money'] = pd.to_numeric(df['money'], errors='coerce')


# Compare Monthly Sales & Transactions

monthly_summary = (
    df.groupby('month')
    .agg(
        total_sales=('money', 'sum'),
        total_transactions=('money', 'count'),
        avg_ticket_size=('money', 'mean')
    )
    .reset_index()
)

# Filter only February and March
monthly_summary = monthly_summary[monthly_summary['month'].isin(['February', 'March'])]

# Sort months for correct order
month_order = ['February', 'March']
monthly_summary['month'] = pd.Categorical(monthly_summary['month'], categories=month_order, ordered=True)
monthly_summary = monthly_summary.sort_values('month')


fig, axes = plt.subplots(1, 2, figsize=(14,6))  # 1 row, 2 columns

# Plot 1: Total Sales
axes[0].bar(monthly_summary['month'], monthly_summary['total_sales'],
            color=['steelblue', 'orange'], edgecolor='black')
axes[0].set_title("Total Sales Comparison", fontsize=14, fontweight='bold')
axes[0].set_xlabel("Month", fontsize=12)
axes[0].set_ylabel("Total Sales (₹)", fontsize=12)
axes[0].grid(axis='y', linestyle='--', alpha=0.6)

# Annotate bars
for i, val in enumerate(monthly_summary['total_sales']):
    axes[0].text(i, val + 300, f"₹{val:,.0f}", ha='center', fontsize=10, fontweight='bold')

#  Plot 2: Average Ticket Size
axes[1].bar(monthly_summary['month'], monthly_summary['avg_ticket_size'],
            color=['skyblue', 'coral'], edgecolor='black')
axes[1].set_title("Average Ticket Size Comparison", fontsize=14, fontweight='bold')
axes[1].set_xlabel("Month", fontsize=12)
axes[1].set_ylabel("Average Ticket Size (₹)", fontsize=12)
axes[1].grid(axis='y', linestyle='--', alpha=0.6)

# Annotate bars
for i, val in enumerate(monthly_summary['avg_ticket_size']):
    axes[1].text(i, val + 0.5, f"₹{val:.2f}", ha='center', fontsize=10, fontweight='bold')

# Adjust layout and main title
fig.suptitle("Customer Behavior Comparison: February vs March", fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

# Insights — February vs March Customer Behavior:

- Total Sales: March recorded higher total sales (₹20,337.64) compared to February (₹16,804.48), showing an increase in revenue and activity.

- Transactions: The number of transactions also rose in March (649 vs 561), suggesting a higher customer footfall or repeat visits.

- Average Ticket Size: The average amount spent per purchase increased slightly from ₹29.95 in February to ₹31.33 in March, indicating customers spent a bit more per order.
-  Conclusion: Customer activity and spending both improved in March, reflecting stronger demand — possibly due to promotions, seasonal factors, or growing customer engagement.

# 4. Cross-Month Comparison:

○ Combine both months and analyze growth trends.

```python

# Coffee Shop Project - Combined Growth Trend (Feb + Mar)
# Clean and prepare data
df.columns = df.columns.str.strip()
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month_name()
df['money'] = pd.to_numeric(df['money'], errors='coerce')

# Filter only February and March
df = df[df['month'].isin(['February', 'March'])]

#  Daily Total Sales for Each Month

daily_sales = df.groupby(['month', 'date'])['money'].sum().reset_index()

# Sort by date for line continuity
daily_sales = daily_sales.sort_values(by='date')

#  Plot 1: Daily Sales Growth Trend

plt.figure(figsize=(12,6))
for month, data in daily_sales.groupby('month'):
    plt.plot(data['date'], data['money'], marker='o', linestyle='-', linewidth=2, label=month)

plt.title("Daily Coffee Sales Growth Trend: February vs March", fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Total Daily Sales (₹)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title="Month")
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()

#  Summary Growth Metrics

monthly_summary = (
    df.groupby('month')
    .agg(
        total_sales=('money', 'sum'),
        total_transactions=('money', 'count'),
        avg_ticket_size=('money', 'mean')
    )
    .reset_index()
    .sort_values('month')
)

# Calculate growth rates
feb_sales = monthly_summary.loc[monthly_summary['month'] == 'February', 'total_sales'].values[0]
mar_sales = monthly_summary.loc[monthly_summary['month'] == 'March', 'total_sales'].values[0]
growth_percent = ((mar_sales - feb_sales) / feb_sales) * 100

print("Monthly Growth Summary:")
print(monthly_summary)
print(f"\n Sales Growth from February to March: {growth_percent:.2f}%")


#  Plot 2: Growth Metrics Comparison (Side-by-Side)

fig, axes = plt.subplots(1, 3, figsize=(16,6))

# --- Total Sales ---
axes[0].bar(monthly_summary['month'], monthly_summary['total_sales'], color=['steelblue', 'orange'], edgecolor='black')
axes[0].set_title("Total Sales", fontsize=13, fontweight='bold')
axes[0].set_ylabel("₹", fontsize=12)
for i, val in enumerate(monthly_summary['total_sales']):
    axes[0].text(i, val + 300, f"₹{val:,.0f}", ha='center', fontsize=10, fontweight='bold')

# --- Total Transactions ---
axes[1].bar(monthly_summary['month'], monthly_summary['total_transactions'], color=['skyblue', 'coral'], edgecolor='black')
axes[1].set_title("Total Transactions", fontsize=13, fontweight='bold')
for i, val in enumerate(monthly_summary['total_transactions']):
    axes[1].text(i, val + 10, f"{val}", ha='center', fontsize=10, fontweight='bold')

# --- Average Ticket Size ---
axes[2].bar(monthly_summary['month'], monthly_summary['avg_ticket_size'], color=['lightgreen', 'darkorange'], edgecolor='black')
axes[2].set_title("Average Ticket Size", fontsize=13, fontweight='bold')
for i, val in enumerate(monthly_summary['avg_ticket_size']):
    axes[2].text(i, val + 0.5, f"₹{val:.2f}", ha='center', fontsize=10, fontweight='bold')

# Layout
fig.suptitle("Coffee Shop Growth Analysis: February → March", fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

```

# Insights — Combined Growth Trend (February → March):




**- Overall Growth:**
There is a strong upward trend from February to March across all key metrics  total sales, transactions, and average ticket size — showing clear business growth.

**-  Sales Growth:**
March’s total sales rose noticeably compared to February, with an approximate increase of ~20–25%, indicating higher revenue and customer activity.

**- Customer Engagement:**
The number of transactions also increased significantly, meaning more customers visited or existing customers made repeat purchases in March.

**- Average Ticket Size:**
The average amount spent per order rose slightly, from around ₹30 in February to ₹31+ in March.
This suggests customers either bought more items per visit or opted for premium drinks.

**-  Daily Trend Observation:**
The daily sales line chart shows more consistency and higher peaks in March, while February had smaller fluctuations.
This points toward steady demand growth and possibly effective business promotions in March.

**-  Conclusion:**
March outperformed February in every major aspect — sales, footfall, and customer spending, signaling positive momentum and improving business performance.

○ Track changes in payment preferences.

```python
# Clean and prepare data
df.columns = df.columns.str.strip()
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month_name()
df['cash_type'] = df['cash_type'].str.title()  # standardize (Card, Cash, Upi)


#  Count payment methods per month

payment_trends = (
    df.groupby(['month', 'cash_type'])
    .size()
    .reset_index(name='transactions')
)

# Pivot for easier plotting
payment_pivot = payment_trends.pivot(index='month', columns='cash_type', values='transactions').fillna(0)

# Sort by chronological order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
payment_pivot = payment_pivot.reindex(month_order).dropna(how='all')


#  Plot : Grouped Bar Chart (Clear Month Labels)

plt.figure(figsize=(10,6))
payment_pivot.plot(kind='bar', ax=plt.gca(), edgecolor='black')

plt.title("Payment Preferences by Month", fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Transactions", fontsize=12)
plt.xticks(rotation=45, ha='right')  # ensures full month names are visible
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title="Payment Type")
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()

#  Summary

print("Payment Trends Summary:")
print(payment_pivot)
```

# Insights — Payment Preferences by Month:

- Dominant Payment Mode:
Across all months, card payments clearly dominate — most customers prefer digital card transactions over cash or UPI.

- Cash Usage:
Cash transactions are very limited, showing that customers rely more on cashless payments, which is typical for cafés with tech-savvy or working professionals.

- UPI Payments:
UPI is either missing or minimally used, indicating it may not be widely adopted or perhaps not offered as a payment option during this period.

- Monthly Pattern:
While total transactions increase slightly in March, the proportion of each payment mode remains consistent, suggesting stable customer behavior in terms of how they pay.

- Business Insight:
The café’s customers show a strong preference for digital payments, especially cards, meaning maintaining fast, reliable digital payment systems should remain a top operational priority.

** In short:**

Card payments lead by a wide margin, while cash and UPI remain minimal — showing a consistent trend toward cashless, convenient transactions month after month.

```python
# Clean and prepare data
df.columns = df.columns.str.strip()
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month_name()
df['money'] = pd.to_numeric(df['money'], errors='coerce')


#  Aggregate Total Sales by Coffee and Month

coffee_sales = (
    df.groupby(['month', 'coffee_name'])['money']
    .sum()
    .reset_index()
)

# Keep only relevant months (February, March)
coffee_sales = coffee_sales[coffee_sales['month'].isin(['February', 'March'])]

# Pivot table for side-by-side comparison
coffee_pivot = coffee_sales.pivot(index='coffee_name', columns='month', values='money').fillna(0)

# Calculate growth percentage
coffee_pivot['Growth_%'] = ((coffee_pivot['March'] - coffee_pivot['February']) / coffee_pivot['February']) * 100

# Sort by March sales (to see which grew)
coffee_pivot = coffee_pivot.sort_values(by='March', ascending=False)


#  Display Pivot Table Output

print("\n Month-over-Month Coffee Sales Comparison Table (₹ & % Growth):\n")
print(coffee_pivot[['February', 'March', 'Growth_%']].round(2).to_string())


#  Plot 1: Side-by-Side Bar Chart (Month-over-Month Sales)

coffee_pivot[['February', 'March']].plot(kind='bar', figsize=(12,6), edgecolor='black')

plt.title("Month-over-Month Coffee Sales Comparison (February vs March)", fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Coffee Type", fontsize=12)
plt.ylabel("Total Sales (₹)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title="Month")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()


#  Plot 2: Coffee Growth Trend (Line Plot)

plt.figure(figsize=(12,6))
for coffee in coffee_pivot.index[:5]:  # top 5 coffees
    plt.plot(['February', 'March'], coffee_pivot.loc[coffee, ['February', 'March']],
             marker='o', linewidth=2, label=coffee)

plt.title("Top 5 Coffees - Month-over-Month Growth Trend", fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Sales (₹)", fontsize=12)
plt.legend(title="Coffee Type")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```
